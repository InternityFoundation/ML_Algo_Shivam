```python
from sklearn.preprocessing import StandardScaler
data2['normAmount'] = StandardScaler().fit_transform(data2['Amount'].reshape(-1, 1))
data2= data2.drop(['Time','Amount'],axis=1)
data2.head()
```

```python
X=data2.iloc[:,data2.columns!="Class"]
y=data2.iloc[:,data2.columns=="Class"]
```

```python
X.head()
```

```python
# as fraud transaction very less so we use undersampling technique
number_records_fraud = len(data2[data2.Class == 1])
fraud_indices = np.array(data2[data2.Class == 1].index)

normal_indices = data2[data2.Class == 0].index


random_normal_indices = np.random.choice(normal_indices, number_records_fraud, replace = False)
random_normal_indices = np.array(random_normal_indices)

under_sample_indices = np.concatenate([fraud_indices,random_normal_indices])

under_sample_data = data2.iloc[under_sample_indices,:]

X_undersample = under_sample_data.ix[:, under_sample_data.columns != 'Class']
y_undersample = under_sample_data.ix[:, under_sample_data.columns == 'Class']
```

```python
from sklearn.cross_validation import train_test_split

# Whole dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)


# Undersampled dataset
X_train_undersample, X_test_undersample, y_train_undersample, y_test_undersample = train_test_split(X_undersample
                                                                                                   ,y_undersample
                                                                                                   ,test_size = 0.3
                                                                                                   ,random_state = 0)
```

```python
#df3 = pd.DataFrame(np.random.randn(10, 5),columns=['a', 'b', 'c', 'd', 'e'])
#df3
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(y_pred)
dff=pd.DataFrame(data=y_pred,columns=['a'])
dff['a'].value_counts()
```

```python
classifier.fit(X_train_undersample, y_train_undersample)

# Predicting the Test set results
y_pred = classifier.predict(X_test_undersample)
print(y_pred)
```
